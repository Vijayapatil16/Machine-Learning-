**Objective:** Implement k‑Nearest Neighbour algorithm using Python ML libraries and analyze predictions.

## Q1: Import Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print('Libraries imported')
```

## Q2: Load Iris Dataset

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
columns = ['sepal_length','sepal_width','petal_length','petal_width','class']
df = pd.read_csv(url, names=columns)
df.head()
```

## Q3: Explore Dataset

```python
print('Shape:', df.shape)
print('Classes:', df['class'].unique())
df.describe()
```

## Q4: Split Features (X) and Target (y)

```python
X = df.iloc[:, 0:4]
y = df['class']
X.head()
```

## Q5: Train‑Test Split (70:30)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print('Training size:', X_train.shape)
print('Testing size:', X_test.shape)
```

## Q6: Build kNN Model (k = 3)

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
print('Model trained.')
```

## Q7: Make Predictions

```python
y_pred = knn.predict(X_test)
list(y_pred)[:10]
```

## Q8: Show Correct and Wrong Predictions

```python
results = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})
results['Status'] = results.apply(lambda r: 'Correct' if r.Actual == r.Predicted else 'Wrong', axis=1)
results
```

## Q9: Evaluate Model Performance

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
print('\nConfusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('\nClassification Report:\n', classification_report(y_test, y_pred))
```
